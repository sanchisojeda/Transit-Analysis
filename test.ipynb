Lectura del archivo

```python
import glob, os
os.chdir("/Users/mariona/Desktop/verano2015")
for file in glob.glob("LCfluxesepic*.txt"):
    data = pd.read_table(file, sep='\s+', header = None, names=['time', 'flux','other'])
    timeLC = data['time']
    fluxLC = data['flux']
    error = np.std(fluxLC)
    meanFlux = np.mean(fluxLC)
    goodP = fluxLC<=meanFlux+3*error
    plt.figure(1)
    plt.plot(timeLC[goodP], fluxLC[goodP],'.')
    plt.xlabel('BJD-2454900', fontsize=12)
    plt.ylabel('Flux', fontsize=12)
```

```python
minTime = min(timeLC) 
maxTime = max(timeLC)

def my_range(start, end, step):
    while start <= end:
        yield start
        start += step
        
minepoch = -10
maxepoch = 20
numepochs = maxepoch-minepoch+1

transitLoc = np.empty(numepochs)

for x in my_range(minepoch, maxepoch, 1):
    transit = t0+x*period
    transitLoc[x-minepoch]=transit
    
goodTrans = (transitLoc<=maxTime) & (transitLoc>=minTime) # Returns boolean
transits = transitLoc[goodTrans] # Stores transit times
```

```python
print transits
num_transits = len(transits)
plt.figure(2)
plt.plot(timeLC[goodP], fluxLC[goodP],'.')
plt.xlabel('BJD-2454900', fontsize=12)
plt.ylabel('Flux', fontsize=12)
for i in range(0,num_transits):
    plt.axvline(transits[i],color='r',ls='dashed')
```

```python
idx_transits = np.empty(len(transits))
x=timeLC[goodP]
y=fluxLC[goodP]

for m in range(0,len(transits)):
    idx_transits[m]=(np.abs(timeLC[goodP] - transits[m])).argmin()

print idx_transits # Prints the approximate x-location of the transits. 

origin=np.empty(len(transits))
final=np.empty(len(transits))
window = 12

for k in range(0,len(transits)):
    origin[k]=int(round(idx_transits[k]-window))
    final[k]=int(round(idx_transits[k]+window))
    
print transits
```

```python
nrows=2*int(window)
ncols=2*len(transits)

flux = np.empty(nrows)
time = np.empty(nrows)
mmatrix=np.zeros((nrows,ncols))

i=0
for k in range(0,len(transits)):
    st=int(origin[k])
    fn=int(final[k])
    time=x[st:fn]
    flux=y[st:fn]
    mmatrix[:,i]=time-transits[k]
    mmatrix[:,i+1]=flux
    i=i+2

plt.figure(3)
plt.subplot(2,2,1) # First Transit
x_1=mmatrix[:,0]
y_1=mmatrix[:,1]
plt.xlim(xmax = mmatrix[nrows-1,0], xmin = mmatrix[0,0])
plt.plot(x_1,y_1,color='black')

plt.subplot(2,2,2) #Second Transit
x_2=mmatrix[:,2]
y_2=mmatrix[:,3]
plt.xlim(xmax = mmatrix[nrows-1,2], xmin = mmatrix[0,2])
plt.plot(x_2,y_2,color='black')

plt.subplot(2,2,3) #Third Transit
x_3=mmatrix[:,4]
y_3=mmatrix[:,5]
plt.xlim(xmax = mmatrix[nrows-1,4], xmin = mmatrix[0,4])
plt.plot(x_3,y_3,color='black')

plt.subplot(2,2,4) #Fourth Transit
x_4=mmatrix[:,6]
y_4=mmatrix[:,7]
plt.xlim(xmax = mmatrix[nrows-1,6], xmin = mmatrix[0,6])
plt.plot(x_4,y_4,color='black')
```
